```python
import pandas as pd
import numpy as np
import requests
from time import sleep
import random 
from bs4 import BeautifulSoup
from collections import defaultdict
import re
import math
import pickle
```

```python

#http://web.mta.info/developers/turnstile.html#http:/ 
#get data data from 12/30/2017 to present
def getData(years):
    merged_dfs = []    
    for year in years:
        print(f'downloading: {year}')
        csv_file = pd.read_csv(f'nytimes_meta_{year}.csv')
        merged_dfs.append(csv_file)
    return pd.concat(merged_dfs, ignore_index = True) #resets index in order to avoid multiples of same index
```

```python
#pass datasets selected
years = []
for year in range(1981,2019):
    years.append(year)
```

```python

df = getData(years)
df
```
